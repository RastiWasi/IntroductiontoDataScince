```python
df = pd.DataFrame()
for i in range(500):
  response = requests.get('http://www.omdbapi.com/?i=tt3896198&apikey=99a17e8d')
  call_df = pd.DataFrame(response.json(), columns=['title', 'Metascore', 'imdbRating', 'imdbID', 'imdbVotes', 'BoxOffice', 'Runtime'], index=[1, 2, 3, 4, 5])
  df = pd.concat([call_df, call_df])
```
